# NB1 — Data Exploration & Understanding
This notebook introduces exploratory data analysis with the Titanic dataset.

## Learning goals
- inspect structure and data types
- classify variables by type
- compute descriptive statistics
- visualize distributions and relationships
- identify basic data quality issues


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid')

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
df = pd.read_csv(url)
print('Shape:', df.shape)
df.head()
```

## Dataset overview


```python
df.info()
```

```python
df.describe(include='all').T
```

## Attribute type classification
A practical classification for this course is shown below.


```python
attribute_types = {
    'PassengerId': 'Identifier',
    'Survived': 'Binary / nominal target',
    'Pclass': 'Ordinal',
    'Name': 'Text / identifier-like',
    'Sex': 'Nominal',
    'Age': 'Continuous',
    'SibSp': 'Discrete',
    'Parch': 'Discrete',
    'Ticket': 'Identifier-like',
    'Fare': 'Continuous',
    'Cabin': 'Nominal / high missingness',
    'Embarked': 'Nominal'
}
pd.DataFrame.from_dict(attribute_types, orient='index', columns=['Type'])
```

## Missing values


```python
missing = df.isna().sum().sort_values(ascending=False)
missing_pct = (missing / len(df) * 100).round(2)
pd.DataFrame({'missing_count': missing, 'missing_pct': missing_pct}).query('missing_count > 0')
```

## Descriptive statistics


```python
numeric_cols = ['Age', 'Fare', 'SibSp', 'Parch']
df[numeric_cols].agg(['mean', 'median', 'std', 'min', 'max']).T
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
sns.histplot(df['Age'].dropna(), bins=30, kde=True, ax=axes[0])
axes[0].set_title('Age distribution')
sns.histplot(df['Fare'].dropna(), bins=30, kde=True, ax=axes[1])
axes[1].set_title('Fare distribution')
plt.tight_layout()
plt.show()
```

```python
corr = df.select_dtypes(include=np.number).corr(numeric_only=True)
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation heatmap')
plt.show()
```
